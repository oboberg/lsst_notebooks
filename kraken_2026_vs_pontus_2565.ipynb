```python
import lsst.sims.maf.metricBundles as mb
import lsst.sims.maf.plots as plots
import lsst.sims.maf.runComparison as rc
```

Here we wil compare 'kraken_2026' and 'pontus_2565' using the new standardComp class.
Before running the following cells make sure you have the following in the same working directory

 - kraken_2026/all
 - pontus_2565/all
 
You should also make a directory to hold the comparison output tables and figures.

~~~
mkdir kraken2026_pontus2565_comp
~~~

This new class is currently on the `feature/standardComp` branch of maf.

Producing the nice markdown tables from pandas required installing a new package called `tabulate`. You can install it with `pip install tabulate`

```python
runList=['kraken_2026','pontus_2565']
```

```python
comp = rc.RunComparison(baseDir='.',runlist=runList)
```

```python
all_dict = comp.buildMetricDict(subdir='all')
```

The next cell will take several minutes because it is collecting and matching all of the metrics in the `all` directory.

```python
with warnings.catch_warnings():
    warnings.simplefilter('ignore')
    comp.addSummaryStats(all_dict)
```

```python
run_comp = rc.standardComp(compOutDir='kraken2026_pontus2565_comp',
                        mafSubDir='all',
                        compObject=comp,
                        runList=['kraken_2026','pontus_2565'])
```

The `generateAll` method will make all of the markdown tables, sky maps, and histograms.

```python
run_comp.generateAll(showTables=True, plotMaps=True, plotHists= True)
```

After running the `generateAll` method there is a new attribute in the class called `comboComp`.
It is a datafram of all of the summary stats.

```python
run_comp.comboComp
```

Finally we can use `mdMaker` method to generate the main `README.md` file that has all of the tables and figures.

```python
run_comp.mdMaker()
```

After running everything you should you should see the following in `kraken2026_pontus2565_comp`

- figures/
- README.md

You can now upload the main directory (`kraken2026_pontus2565_comp`) to GitHub and direct people to the README.md file.
